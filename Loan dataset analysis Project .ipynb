```python
df_5['emp_length']=df_5['emp_length'].str.extract('(\d+)').astype(float)
df_5 #We use the str.extract() method with the regular expression (\d+) to extract one or more digits from each string. This regular expression captures the numerical value.
```

```python
df_5['risky_loan_applicant'] = df_5.apply(lambda row: '0' if row['loan_amnt'] <= row['funded_amnt'] else '1', axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Plot 'loan_status' against 'grade'
sns.countplot(x='grade', hue='loan_status', data=df_5, ax=axes[0])
axes[0].set_title('Loan Status vs. Grade')

# Plot 'loan_status' against 'term'
sns.countplot(x='term', hue='loan_status', data=df_5, ax=axes[1])
axes[1].set_title('Loan Status vs. Term')

# Plot 'loan_status' against 'verification_status'
sns.countplot(x='verification_status', hue='loan_status', data=df_5, ax=axes[2])
axes[2].set_title('Loan Status vs. Verification Status')
```

```python
df_5['emp_lenght']=df_5['emp_lenght'].astype(float)
```

```python
def recode_emp_len(emp_lenght):
    if emp_lenght <= 1:
        return 'fresher'
    elif emp_lenght > 1 and emp_lenght < 3:
        return 'junior'
    elif emp_lenght >= 3 and emp_lenght < 7:
        return 'senior'
    else:
        return 'expert'

# Apply the custom function to 'emp_len' and create a new column 'emp_category'
df_5['emp_category'] = df_5['emp_lenght'].apply(recode_emp_len)
df_5
```

```python
grade_loan_sum = df_5.groupby('grade')['loan_amnt'].sum().reset_index()

# Plot the distribution of 'loan_amnt' using a pie plot
plt.figure(figsize=(8, 6))
plt.pie(grade_loan_sum['loan_amnt'], labels=grade_loan_sum['grade'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Loan Amount by Grade')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie plot
plt.show()
```
